```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
qbs = pd.read_csv('quarterbacks.csv')
game_logs = pd.read_csv('game_logs.csv')
qbs.head()
```

```python
qbs['team'] = qbs['team'].apply(lambda x: 'Washington Commanders' if x == 'Washington' else x)
```

```python
game_logs['Winner/tie'] = game_logs['Winner/tie'].apply(lambda x: 'Washington Commanders' if x == 'Washington Football Team' else x)
game_logs['Loser/tie'] = game_logs['Loser/tie'].apply(lambda x: 'Washington Commanders' if x == 'Washington Football Team' else x)
```

```python
game_logs.head()
```

```python
def get_year(date_str):
    if date_str[5:7] == '01':
        return int(date_str[:4]) - 1
    else:
        return int(date_str[:4])

game_logs['year'] = game_logs['Date'].apply(get_year)
game_logs.head()
```

```python
game_logs['home_away_indicator'].value_counts()
```

```python
game_logs.shape
```

```python
def get_home_team(row):
    if row['home_away_indicator'] == '@':
        return row['Loser/tie']
    else:
        return row['Winner/tie']
    
def get_away_team(row):
    if row['home_away_indicator'] == '@':
        return row['Winner/tie']
    else:
        return row['Loser/tie']

game_logs['home_team'] = game_logs.apply(get_home_team, axis=1)
game_logs['away_team'] = game_logs.apply(get_away_team, axis=1)
game_logs.head()
```

```python
game_logs[((game_logs['home_team'] == 'Washington Commanders') | (game_logs['away_team'] == 'Washington Commanders')) & (game_logs['year'] == 2020)]
```

```python
# drop a bunch of columns from game logs cuz we don't need em
game_logs.drop(columns=['Day', 'Time', 'home_away_indicator', 'Unnamed: 7', 'Pts', 'Pts.1', 'YdsW', 'TOW', 'YdsL', 'TOL'], inplace=True)
game_logs.head()
```

```python
print(game_logs.shape)
game_logs.drop_duplicates(inplace=True)
print(game_logs.shape)
```

```python
game_logs.rename(columns={'Loser/tie': 'loser', 'Winner/tie': 'winner', 'Week': 'week', 'Year': 'year'}, inplace=True)
```

```python
home = qbs.merge(game_logs, left_on=['week', 'year', 'team'], right_on=['week', 'year', 'home_team'], how='left')
```

```python
home.isna().sum()
```

```python
away = qbs.merge(game_logs, left_on=['week', 'year', 'team'], right_on=['week', 'year', 'away_team'], how='left')

away.isna().sum()
```

```python
home.dropna(inplace=True)
away.dropna(inplace=True)

print(home.shape)
print(away.shape)
```

```python
home['home_indicator'] = 1
away['home_indicator'] = 0

home['win_indicator'] = home.apply(lambda x: 1 if x['winner'] == x['team'] else 0, axis=1)
away['win_indicator'] = away.apply(lambda x: 1 if x['winner'] == x['team'] else 0, axis=1)
```

```python
qbs.shape
```

```python
qbs = pd.concat([home, away], ignore_index=True)

qbs.shape
```

```python
from geopy.geocoders import Nominatim

def get_lat_long(location):
    for i in range(3):
        try:
            geolocator = Nominatim(user_agent="nfl")
            location = geolocator.geocode(location)
            return location.latitude, location.longitude
        except:
            pass
    return None, None
```

```python
# for the unique values in the location column, get the lat and long
locs = qbs['location'].unique()
lat_longs = {loc: get_lat_long(loc) for loc in locs}

lat_longs
```

```python
# add the lat and long to the dataframe
qbs['lat'] = qbs['location'].map(lambda x: lat_longs[x][0])
qbs['long'] = qbs['location'].map(lambda x: lat_longs[x][1])
```

```python
qbs.head()
```

```python
from meteostat import Point, Daily
from datetime import datetime

def get_historical_temp(lat, long, date):
    try:
        point = Point(lat, long)
        data = Daily(point, start=date, end=date)
        data = data.fetch()
        return data['tavg'].values[0]
    except Exception as e:
        print(e)
        return np.nan
```

```python
import warnings
from tqdm import tqdm
warnings.filterwarnings('ignore')

tqdm.pandas()

# convert date to datetime
qbs['Date'] = pd.to_datetime(qbs['Date'])

# get historical temp for each game
qbs['historical_temp'] = qbs.progress_apply(lambda x: get_historical_temp(x['lat'], x['long'], x['Date']), axis=1)
```
